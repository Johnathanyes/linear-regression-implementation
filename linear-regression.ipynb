```python
"""
This program is an implementation of the Linear Regression algorithm
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

x_train = np.array([1.0, 2.0, 3.0])   #features
y_train = np.array([300.0, 500.0, 700.0])   #target value
```

```python
def cost_function(x, y, w, b):
    m = x.shape[0]
    cost = 0

    for i in range(m):
        eq = w * x[i] + b
        cost += ((eq - y[i]) ** 2)
    cost = cost / (2 * m)
    return cost
```

```python
def compute_gradient(x, y, w, b):
    m = x.shape[0]
    dj_dw = 0
    dj_db = 0

    for i in range(m):
        f_wb = w * x[i] + b
        temp_dj_dw = (f_wb - y[i]) * x[i]
        temp_dj_db = (f_wb - y[i])
        dj_dw += temp_dj_dw
        dj_db += temp_dj_db
    dj_dw = dj_dw / m
    dj_db = dj_db / m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w_in, b_in, a_in, num_iters):
    w = w_in
    b = b_in
    a = a_in
    iterations = num_iters

    for i in range(iterations):
        dj_dw, dj_db = compute_gradient(x, y, w, b)
        w = w - a * dj_dw
        b = b - a * dj_db
    return w, b
```

```python
w_in = 0
b_in = 0
num_iterations = 10000
a = 0.01

w, b = gradient_descent(x_train, y_train, w_in, b_in, a, num_iterations)
prediction = np.arange(0, 4) * w + b
plt.scatter(x_train, y_train)
plt.plot(prediction)
print(f"{w} * x + {b}")
```
